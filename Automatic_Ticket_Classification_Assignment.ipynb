<b><font color ='Black' style=font-family:Verdana;font-size:18px>Text preprocessing</font><br>

<font color ='Black' style=font-family:Verdana;font-size:18px>Prepare the text for topic modeling</font><br>

  - <font color ='brown' style=font-family:Verdana;font-size:13px>Make the text lowercase.</font>

  - <font color ='brown' style=font-family:Verdana;font-size:13px>Remove text in square brackets</font>

  - <font color ='brown' style=font-family:Verdana;font-size:13px> Remove punctuation. & Remove words containing numbers</font>
  
  

<font color ='Black' style=font-family:Verdana;font-size:18px>Once you have done these cleaning operations you need to perform the following:</font><br>

  - <font color ='brown' style=font-family:Verdana;font-size:13px>Lemmatize the texts.</font>

  - <font color ='brown' style=font-family:Verdana;font-size:13px>Extract the POS tags of the lemmatized text and remove all the words which have tags other than NN[tag == "NN"].m</font>


```python
# Write your function here to clean the text and remove all the unnecessary elements.
def text_preprocessing(text):
    #step 1: text to lower casae
    text=text.lower()
    
    # step 2 : removing any leading and trailing spaces
    text=text.strip()
    
    # step 3: Removing text within square brackets
    text=re.sub(r"[\[].*?[\]]","",text).strip()
    
    # step 4 removing punctions:
    text = re.sub(r"[^\w\s]","",text).strip()
    
    # Removing words containing numbers
    text = re.sub("\S*\d\S*","",text).strip()   
    
    return text
```

```python
#Write your function to Lemmatize the texts
df['complaints_cleaned']=df['complaint_what_happened'].apply(text_preprocessing)
```

```python
stopwords_list=stopwords.words('english')
wordnet_lem=WordNetLemmatizer()
def lemitizer(text):
    
    tokens=word_tokenize(text)
    tokens=[i for i in tokens if i not in stopwords_list]
    proc_token=[]
    # Using nltk pos tagger which may not capture the context like Spacy's POS tagger. However, it should be good enough for use here.
    for i,j in pos_tag(tokens):
        try:
            proc_token.append(wordnet_lem.lemmatize(i,j=j[0].lower()))
        except:
            proc_token.append(wordnet_lem.lemmatize(i))
    lemmatized_text=' '.join(proc_token)
    return lemmatized_text
        
```



<font color ='brown' style=font-family:Verdana;font-size:13px>It is more likely the words of interest for Topic Modelling are nouns and proper nouns. Words with other POS tags are unlikely to be of much use to us. Hence, the code below creates a new column which includes only words with POS tags as nouns. </font>

```python
nlp=spacy.load('en_core_web_sm')
from tqdm import tqdm
pos_tag=[] # parts of speech tag
nouns=[]
for i in tqdm(df_clean['lemmatized_complaints']):
    parts_of_speech=''
    noun=''
    complaint=nlp(i)
    for j in complaint:
        parts_of_speech+=j.pos_+' '
        if j.pos_ in str(('NOUN','PROPN')):
            noun+= (j.lemma_.lower())+ ' '
    pos_tag.append(parts_of_speech.strip())
    nouns.append(noun.strip())
```

```python
df_clean['pos_tags_lemmatized']=pd.Series(pos_tag,index=df_clean.index)
df_clean['complaints_pos_removed'] = pd.Series(nouns, index=df_clean.index)

#Reassigning index of cleaned dataframe
df_clean.reset_index(drop=True, inplace = True)
```

```python
#Write your code here to find the top 30 bigram frequency among the complaints in the cleaned datafram(df_clean). 
def get_top_bigram(text,n=30):
    vector=CountVectorizer(ngram_range=(2,2),stop_words=('english')).fit(text)
    bag_of_words=vector.transform(text)
    sum_of_words=bag_of_words.sum(axis=0)
    word_freq=[ (word, sum_of_words[0,idx]) for word ,idx in vector.vocabulary_.items()  ]
    word_freq =sorted(word_freq, key = lambda x: x[1], reverse=True)
    return word_freq[:n]
```

```python
from sklearn import metrics

def model_evaluation(y_actual,y_pred):
    
    #confusion metrix
    confusion=metrics.confusion_matrix(y_actual,y_pred)
    
    #metrix calculation
    accuracy=metrics.accuracy_score(y_actual,y_pred)
    clf_report=metrics.classification_report(y_actual,y_pred)
    
    
    
    print('ACCURACY :',str(accuracy))
    print('THE CONFUSION METRIX IS :\n',confusion)
    print('\n CLASIFICATION REPORT \n :',clf_report)
    return accuracy
    

```

<font color ='brown' style=font-family:Verdana;font-size:15px>Predicting topics via Logistic Regression on custom text</font>

```python
Topic_names =   {
                0 : "Bank account services",
                1 : "Credit Card/Prepaid Card",
                2 : "Mortgages/loans",
                3 : "Theft/Dispute reporting",
                4 : "Others"
                }
```

```python
# Function to first vectorize the custom text for use with our model. The function then predicts the appropriate topic for the text.
def predict_complaint(sentence, model=logreg):
    
    # Vectorizing the text to apply the model. Using the same vectorizer which was fit on the training data to maintain parity.
    vect_custom = vectorizer.transform(pd.Series(sentence))
    word_vect_custom = tfidf_transform.transform(vect_custom)

    # Converting to a dataframe
    word_vect_custom_df = pd.DataFrame(word_vect_custom.toarray(),columns=vectorizer.get_feature_names_out())

    # Making predictions
    custom_pred = model.predict(word_vect_custom_df)
    
    # Printing predicted output. Using the Topic_names dictionary mapping defined earlier in the notebook.
    print("The text can be classified under topic number {0}, which belongs to the following category: {1}".format(custom_pred[0],Topic_names[custom_pred[0]]))
```
